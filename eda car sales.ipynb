```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt                                    
import seaborn as sns                                              
%matplotlib inline
sns.set()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/car_sales.csv",encoding='latin-1')
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.replace({'engType': 'Other', 'body': 'other','price': 0, 'mileage': 0}, np.nan, inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df['price']=df['price'].fillna(0,inplace=False)
df['engV']=df['engV'].fillna(0,inplace=False)
df['mileage']=df['mileage'].fillna(0,inplace=False)
df['body']=df['body'].fillna(0,inplace=False)
```

```python
df['drive']=df['drive'].ffill
df['engType']=df['engType'].ffill
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.car.value_counts().head(10).plot.bar()
plt.title("Top 10 car brands on sale")
```

```python
df['model'].value_counts().head(10).plot.bar()
```

```python
fig_size = (10,10)
fig, ax = plt.subplots(figsize=fig_size)
sns.heatmap(df.corr(),annot=True, linewidths=.5,ax=ax)
plt.title("Heatmap for Highest correlated features for Carsale datset")
```

```python
fig_size = (10,5)
fig, ax = plt.subplots(figsize=fig_size)
sns.countplot(y='body', data=df, orient='h', hue='registration',ax=ax)
plt.title("Most preferred Engine Type used in 1953-2016")
```

```python
fig_size = (15, 10)
fig, ax = plt.subplots(figsize=fig_size)
sns.lineplot(data=df, y='price', x='year', hue='drive',ax=ax)
plt.title("year - price lineplot (1950 - 2010)")
```

```python
sns.pairplot(df, hue='engType', palette="viridis", height=3)
```
